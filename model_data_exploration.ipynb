```python
import pandas as pd

df = pd.read_csv('processed_data.csv')

df = df[df['HHFAMINC'] > 0].reset_index()

display(df)
print(df.columns)
```

```python
numerical = [
            'TAXISERVICE', 'ESCOOTERUSED', 'PTUSED', 'MCTRANSIT', 'WALKTRANSIT',
            'BIKETRANSIT', 'EMPLOYMENT2', 'WTHHFIN', 'WTHHFIN5D', 'WTHHFIN2D',
            'CNTTDHH', 'PPT517', 'YOUNGCHILD', 'RESP_CNT','WRKCOUNT', 'NUMADLT',
            'RIDESHARE22', 'DRVRCNT', 'HHSIZE', 'HHVEHCNT'
            ]

categorical = [
                'HighestEDUC', 'HOMEOWN', 'HOMETYPE', 'TRAVDAY', 
                'RAIL', 'CENSUS_D', 'CENSUS_R', 'HH_HISP',
                'FLAG100', 'HHFAMINC', 'HHFAMINC_IMP', 'HH_RACE', 'CDIVMSAR', 
                'HHRELATD', 'LIF_CYC', 'MSACAT', 'MSASIZE',
                'URBAN', 'URBANSIZE', 'URBRUR', 'URBRUR_2010', 
               ]
```

```python
for i in range(len(df)):
    if (df['HH_RACE'][i] == 97):
        df['HH_RACE'][i] = 7

for i in range(len(df)):
    df['CDIVMSAR'][i] = df['CDIVMSAR'][i] - 10
```

```python
import numpy as np
import torch

df_categorical = {}
freature_lengths = {}

for colu in categorical:
    df_categorical[colu] = []

for i in range(len(df)):
    for colu in categorical:
        freature_lengths[colu] = df[colu].max()
        df_categorical[colu].append(torch.zeros(df[colu].max()))
        vec = torch.zeros(df[colu].max())

        vec[df[colu][i] - 1] = 1
        df_categorical[colu][i] = vec

df_categorical = pd.DataFrame(df_categorical)
```

```python
display(df_categorical)
df_categorical.columns
```

```python
df_numerical = {}

for colu in numerical:
    df_numerical[colu] = []

for i in range(len(df)):
    for colu in numerical:
        df_numerical[colu].append(df[colu][i])

df_numerical = pd.DataFrame(df_numerical)
```

```python
display(df_numerical)
df_numerical.columns
```

```python
encoded = pd.concat([df_categorical, df_numerical], axis=1)
encoded
```

```python
# Convert object type columns to float64

lst = []
for i in range(len(df)):
    lst.append([])
    for col in df_categorical.columns:
        lst[i] = lst[i] + df_categorical[col][i].tolist()

for i in range(len(df)):
    for col in df_numerical.columns:
        lst[i].append(df[col][i])


features = torch.tensor(lst, dtype=torch.float32)
labels = torch.tensor(df['GASUSGE'], dtype=torch.float32)
```

```python
feature_names = []

for col in df_categorical.columns:
    feature_names = feature_names + [col + str(i) for i in range(1, freature_lengths[col] + 1)]

feature_names = feature_names + df_numerical.columns.tolist()
```

```python
import matplotlib.pyplot as plt
import numpy as np

from sklearn.linear_model import RidgeCV

ridge = RidgeCV(alphas=np.logspace(-6, 6, num=5)).fit(features, labels)
importance = np.abs(ridge.coef_)
plt.bar(height=importance, x=feature_names)
plt.title("Feature importances via coefficients")
plt.show()
```
